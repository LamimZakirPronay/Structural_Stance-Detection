# Unsupervised Structural Embedding for Stance Detection

In this project, we propose an unsupervised approach to propagate the stance of posts in online discussions. We use a new algorithm with inspiration from  heat diffusion graph network -based approach  to propagate the stance of posts to their neighboring posts in a discussion graph. Our approach does not require labeled data, which makes it ideal for scenarios where labeled data is limited or unavailable. Our algorithm is able to handle cyclic conversions which makes it a best suit for platforms like Facebook or Twitter where actual debate takes place.

```python
def edgeweight(charlie_node_id, david_node_id, net):
    for edge in net.edges:
        if edge['from'] == charlie_node_id and edge['to'] == david_node_id:
            edge_weight = edge.get('value', 0)
            break
    else:
        edge_weight = 0
    return edge_weight
```

## Chapter 2

Propagation of Stance

We have choosen the accuracy as our performance measure and also to understand amount of loss we have used cross entropy

```python

def cross_entropy(y_true, y_pred):
    epsilon = 1e-15  # to prevent division by zero
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)  # clip the values to avoid numerical instability
    ce = -np.sum(y_true * np.log(y_pred))  # calculate the cross-entropy
    return ce
```

## The Heat Propagataion Algorithim

```python


 from sklearn.metrics import f1_score

def Heat_Propagation(alpha1, threshold1):
          alphaaa=[]
          thresholddd=[]
          # read the CSV file into a pandas DataFrame
          data = pd.read_csv('file1.csv')

          # filter the DataFrame to only include rows with the specified discussion_id
          
          # create a directed graph
          G = nx.DiGraph()

          # add edges and weights to the graph
          for _, row in data.iterrows():
              source = row['Child']
              target = row['Parent']
              if target != 0:
                  weight = row['score']
                  if G.has_edge(target, source):
                      G[target][source]['weight'] += weight
                  else:
                      G.add_edge(target, source, weight=weight)

          # normalize edge weights
          for edge in G.edges():
              source = edge[0]
              weights = np.array([G.get_edge_data(*edge)['weight']])
              normalized_weights = normalize(weights.reshape(1, -1), norm='l1')
              G[source][edge[1]]['weight'] = normalized_weights[0][0]

          # initialize the node stances
          for node in G.nodes():
              G.nodes[node]['stance'] = 0.0

          # set the source node stance
          source_nodes = [node for node in G.nodes() if G.in_degree(node) == 0]
          for source_node in source_nodes:
              G.nodes[source_node]['stance'] = 1.0

          # heat-based propagation algorithm
          alpha = alpha1
          num_iterations = 10
          for i in range(num_iterations):
              for node in G.nodes():
                  new_stance = 0.0
                  for neighbor in G.predecessors(node):
                      weight = G[neighbor][node]['weight']
                      neighbor_stance = G.nodes[neighbor]['stance']
                      new_stance += weight * neighbor_stance
                  new_stance *= alpha
                  new_stance += (1 - alpha) * G.nodes[node]['stance']
                  G.nodes[node]['stance'] = new_stance

          # label the nodes based on their stance
          
          threshold = threshold1
          for node in G.nodes():
              stance = G.nodes[node]['stance']
              if stance > threshold:
                  G.nodes[node]['Ground'] = 1
              elif stance<threshold:
                  G.nodes[node]['Ground'] = -1
              
              else:
                  G.nodes[node]['Ground'] = 0

          # load the ground truth data
          truth = pd.read_csv('file1.csv')

          # get the predicted and true labels
          predicted_labels = []
          true_labels = []
          for _, row in truth.iterrows():
              Child = row['Child']
              if Child in G.nodes():
                  predicted_labels.append(G.nodes[Child]['Ground'])
                  true_label = row['Ground']
                  if true_label == 'supported':
                      true_labels.append(1)
                  elif true_label == 'disputed':
                      true_labels.append(-1)
                  else:
                      true_labels.append(-1)


          
          
          #print("With the alpha values of "+str(alpha)+"and the threshold value is "+str(threshold))
          alphaaa.append(alpha)
          f1_accuracy = f1_score(true_labels, predicted_labels, average='macro')
          
          return f1_accuracy 
            
            
            #print("F1 Accuracy:", f1_accuracy)
          #print("The Cross Entropy Loss Value is "+str(cross_entropy(true_labels,predicted_labels)))
            
```
